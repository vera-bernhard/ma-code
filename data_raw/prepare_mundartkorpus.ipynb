```python
import os
def count_sentences_in_directory(directory):
    sentence_count = 0
    
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        
        if os.path.isfile(file_path) and filename.endswith(".txt"):
            with open(file_path, "r", encoding="utf-8") as file:
                sentence_count += sum(1 for _ in file)  # Each line is a sentence
    
    return sentence_count
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def count_sentence_lengths(directory):
    sentence_lengths = []

    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)

        if os.path.isfile(file_path) and filename.endswith(".txt"):
            with open(file_path, "r", encoding="utf-8") as file:
                for line in file:
                    tokens = line.strip().split()
                    sentence_lengths.append(len(tokens))
    return sentence_lengths

def plot_sentence_length_histogram(sentence_lengths, bin_size=10, max_length=512):
    """Plots a histogram of sentence lengths using Seaborn."""
    
    sns.set_style("whitegrid")
    plt.figure(figsize=(15, 5))
    sns.histplot(sentence_lengths, bins=range(0, max_length + bin_size, bin_size), kde=True, color="royalblue")

    plt.xlim(0, max_length)
    
    plt.xlabel("Sentence Length (Number of Tokens)")
    plt.ylabel("Frequency")
    plt.title("Sentence Length Distribution")
    plt.xticks(range(0, max_length + bin_size, bin_size))
    
    plt.show()
```

```python
sent_lengths = count_sentence_lengths(data_path)
```

```python
plot_sentence_length_histogram(sent_lengths, bin_size=5, max_length=100)
```
